# IF29 — Labellisation manuelle

## Étape 3 — Après l'EDA, avant le ML

```
users_aggregated.csv (21 var., EDA terminée, 16 features retenues)
        ↓
   Labellisation Excel (ce document)
        ↓
users_labeled_manual.csv (+ label + anomaly_score)  →  notebooks ML
```

**Entrée :** `users_aggregated.csv` — 643 124 profils, 21 variables, sans label.  
**Sortie :** `users_labeled_manual.csv` — mêmes profils + 2 colonnes de labellisation.

Aucun code Python ne génère les labels (Excel uniquement).  
Détails : `docs/LABELISATION.md`

## 1. Processus Excel

**Prérequis :** analyse exploratoire terminée (16 features ML retenues sur 21 variables agrégées).

1. Export `users_aggregated.csv` vers Excel
2. Filtres critère par critère (guidés par l'EDA)
3. Inspection visuelle
4. Label atypique si >= 2 critères sur 4
5. Export vers `users_labeled_manual.csv`

## 2. Critères

| # | Règle | Condition |
|---|-------|----------|
| 1 | Amplification | `retweet_ratio >= 0,8` |
| 2 | Spam | `avg_urls >= 1,5` OU `avg_mentions >= 2` |
| 3 | Burst | `active_days = 1` ET `nb_tweets >= 2` |
| 4 | Déséquilibre | `followers_friends_ratio >= 30` |

`anomaly_score` 0-1 : Normal · >= 2 : Atypique (`label = 1`)

## 3. Fichier commun ML

| Colonne | Non supervisé | Supervisé |
|---------|---------------|-----------|
| Features | 16 (retenues après EDA) | 8 hors règles Excel |
| `label` | Évaluation seulement | Cible |
| `anomaly_score` | Exclu | Exclu |

**Pipeline ML :** StandardScaler → ACP (7 comp. non sup. · 5 comp. sup.) → modèles

Voir `docs/LABELISATION.md`
