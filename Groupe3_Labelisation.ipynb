# IF29 — Labellisation manuelle

Documentation du processus Excel. Aucun code Python de génération de labels.

Fichier produit : **`users_labeled_manual.csv`** — utilisé par non supervisé et supervisé.

Détails : `docs/LABELISATION.md`

## 1. Processus Excel

1. Export `users_aggregated.csv` vers Excel
2. Filtres critère par critère
3. Inspection visuelle
4. Label atypique si >= 2 critères sur 4
5. Export vers `users_labeled_manual.csv`

## 2. Critères

| # | Règle | Condition |
|---|-------|----------|
| 1 | Amplification | `retweet_ratio >= 0,8` |
| 2 | Spam | `avg_urls >= 1,5` OU `avg_mentions >= 2` |
| 3 | Burst | `active_days = 1` ET `nb_tweets >= 2` |
| 4 | Déséquilibre | `followers_friends_ratio >= 30` |

`anomaly_score` 0-1 : Normal · >= 2 : Atypique (`label = 1`)

## 3. Fichier commun ML

| Colonne | Non supervisé | Supervisé |
|---------|---------------|-----------|
| Features | 16 MongoDB | 8 hors règles Excel |
